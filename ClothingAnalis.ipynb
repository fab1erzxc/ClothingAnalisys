### 6. Customer Segmentation:



```python
import sklearn
# Select relevant features for segmentation
segmentation_data = data_frame[['Age', 'Purchase Amount (USD)']]

# Scale the features for k-means clustering
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(segmentation_data)
```

```python
from sklearn.cluster import KMeans
import numpy as np

# Using the elbow method to find the optimal number of clusters (k)
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(scaled_data)
    wcss.append(kmeans.inertia_)

# Plot the elbow method graph
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.show()
```

```python
# Choose the optimal number of clusters (k)
k = 3  # You can adjust this based on the elbow method result

# Apply k-means clustering
kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
data_frame['Cluster'] = kmeans.fit_predict(scaled_data)
```

```python
# Plot the customer segmentation
plt.figure(figsize=(10, 6))
for cluster in range(k):
    plt.scatter(data_frame[data_frame['Cluster'] == cluster]['Age'], data_frame[data_frame['Cluster'] == cluster]['Purchase Amount (USD)'],
                label=f'Cluster {cluster + 1}')
plt.title('Customer Segmentation')
plt.xlabel('Age')
plt.ylabel('Purchase Amount (USD)')
plt.legend()
plt.show()
```

### 9. Correlations and Hypotheses:



```python
# Select only numerical columns for correlation analysis
numerical_columns = ['Age', 'Purchase Amount (USD)', 'Review Rating']

# Calculate the correlation matrix for numerical columns
correlation_matrix = data_frame[numerical_columns].corr()

# Visualize the correlation matrix as a heatmap
import seaborn as sns
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
from scipy import stats

# Separate the data into two groups: with and without promo codes
with_promo = data_frame[data_frame['Promo Code Used'] == 'Yes']['Purchase Amount (USD)']
without_promo = data_frame[data_frame['Promo Code Used'] == 'No']['Purchase Amount (USD)']

# Perform a t-test to compare means
t_stat, p_value = stats.ttest_ind(with_promo, without_promo)

# Check the p-value to determine statistical significance
if p_value < 0.05:
    print(f"The difference in purchase amounts between groups is statistically significant (p-value={p_value:.4f}).")
else:
    print(f"The difference in purchase amounts between groups is not statistically significant (p-value={p_value:.4f}).")
```
